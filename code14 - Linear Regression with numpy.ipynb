```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
area = [1.9,2.4,5.3,3.6,2.5,4.2,5.2,6.3,1.2,3.6,4.3,4.1,6.2,3.8,4.5]
price= [210,250,480,390,220,460,490,580,180,350,410,450,580,450,425]
```

```python
x = np.array(area)
y = np.array(price)
m = 2.5
c = 10.9
LR = 0.003

for i in range(500):
    # get predictions
    yhat = m*x + c
    
    # calculate error
    error = np.mean((y-yhat)**2)
    
    # Gradient descent to update/change the values of m and c
    dedm = 2*np.dot((yhat-y),x)
    m = m - LR*dedm
    
    dedc = 2*np.dot((yhat-y),np.ones(y.size))
    c = c - LR*dedc
    
    print(error)
    
```

```python
print(m,c)
```

# linear regression with sklearn

```python
x = np.array(area).reshape(15,1)
y = np.array(price)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# train the model and get optimized model to make predictions
model.fit(x,y)
```

```python
print(model.coef_,model.intercept_)
```

```python
# makeing prediction for area = 2600 sq ft
model.predict([[2.6]])
```
